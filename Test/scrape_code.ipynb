```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.target.com/s?searchTerm=pizza'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='styles__ProductCardItemInfoDiv')

for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('div', class_='h-display-flex').text
        #price = result.find('data-test', class_='current-price').text

        if (title):
            print('-------------')
            print(title)
            #print(price)
    except AttributeError as e:
        print(e)
```
