```python
import os
from bs4 import BeautifulSoup as bs
import urllib
import pandas as pd
import json
```

# Mac N Cheese Page 1

```python
url_1 = urllib.request.urlopen('file:///C:/Users/willt/OneDrive/Desktop/project3/WebScrape/Walmart/mac_n_cheese/cheese_one.html')
soup = bs(url_1)
```

```python
results = soup.find_all('div', class_='h-100 pb1-xl pr4-xl pv1 ph1')


class items: 
  def __init__(self, brand, price):
    self.brand = brand_1
    self.price = price_1

#lists for brands and items
item_list_1 = []
brands_1 = []


for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        price_1 = result.find('div', class_='b black f5 mr1 mr2-xl lh-copy f4-l').text   
        brand_1 = result.find('span', class_='f6 f5-l normal dark-gray mb0 mt1 lh-title').text

        
        if (brand_1 and price_1):
            new_dict_1 = {
                       'brand': brand_1,
                       'price': price_1
                       }
            print('-------------')
            print(brand_1)
            print(price_1)
    except AttributeError as e:
        print(e)

  # appending instances to list 
    item_list_1.append( new_dict_1)
    brands_1.append(brand_1)
```

```python
cheese_one = pd.DataFrame (item_list_1)
cheese_one.head(20)
```
