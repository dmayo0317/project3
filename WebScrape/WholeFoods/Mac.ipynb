```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.wholefoodsmarket.com/search?text=mac+and+cheese'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='w-pie--product-tile__content')

for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        header = result.find('span', class_='w-cms--font-disclaimer').text
        title = result.find('h2', class_='w-cms--font-body__sans-bold').text
        # Identify and return price of listing
        
        if (header and title):
            print('-------------')
            print(header)
            print(title)
    except AttributeError as e:
        print(e)
```
