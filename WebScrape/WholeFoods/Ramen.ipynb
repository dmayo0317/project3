```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
import urllib
import pandas as pd
import json
```

```python
ramenurl = urllib.request.urlopen('file:///C:/Users/Matt/Desktop/Projects/Project-3/project3/WebScrape/WholeFoods/ramen.html')
soup = bs(ramenurl)
```

```python
results = soup.find_all('div', class_='w-pie--product-tile__content')


class items: 
  def __init__(self, brand, title, price):
    self.brand = brand
    self.title = title
    self.price = price

#lists for brands and items
item_list = []
brands = [] 


for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        brand = result.find('span', class_='w-cms--font-disclaimer').text
        title = result.find('h2', class_='w-cms--font-body__sans-bold').text
        price = result.find('span', class_='regular_price').text

        
        if (brand and title and price):
            new_dict = {'brand': brand,
                       'title': title,
                       'price': price
                       }
            print('-------------')
            print(brand)
            print(title)
            print(price)
    except AttributeError as e:
        print(e)

  # appending instances to list 
    item_list.append( new_dict)
    brands.append(brand)
```

```python
wf_ramendf = pd.DataFrame(item_list)

wf_ramendf
```
