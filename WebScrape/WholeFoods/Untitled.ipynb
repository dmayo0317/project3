```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
import urllib
import pandas as pd
import json
```

```python
url = urllib.request.urlopen('file:///C:/Users/Matt/Desktop/Projects/Project-3/project3/WebScrape/Walgreens/walgreens_ramen.html')
soup = bs(url)
```

```python

results = soup.find_all('div', class_='product__text')


class items: 
  def __init__(self, brand, title, price):
    self.brand = brand
    self.title = title
    self.price = price

#lists for brands and items
item_list = []
brands = [] 


for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        brand = result.find('div', class_='brand').text
        title = result.find('strong', class_='description').text
        price = result.find('span', class_='product__price font__sixteen wag-inline-txt').text

        
        if (brand and title and price):
            new_dict = {'brand': brand,
                       'title': title,
                       'price': price
                       }
            print('-------------')
            print(brand)
            print(title)
            print(price)
    except AttributeError as e:
        print(e)

  # appending instances to list 
    item_list.append( new_dict)
    brands.append(brand)
```

```python
wa_ramendf = pd.DataFrame(item_list)

wa_ramendf
```
