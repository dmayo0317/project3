```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://shop.sprouts.com/search?search_term=pizza&sort=rank&category_id=218'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='cell-content-wrapper')

print(results)

for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('div', class_='cell-title-text').text

        # Print results only if title, price, and link are available
        if (title):
            print('-------------')
            print(title)
    except AttributeError as e:
        print(e)
```
